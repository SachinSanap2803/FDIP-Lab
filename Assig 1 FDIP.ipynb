```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("C:/Users/sanap/Downloads/sunset_luxury_home.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
if image is None:
    print("Error: Image not found. Check the path!")
else:
    # Show original image
    plt.figure(figsize=(20, 15))
    plt.subplot(2, 4, 1)
    plt.imshow(image, cmap="gray")
    plt.title("Original Image")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
C = 50 
bright_improve = cv2.add(image, C)
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 2)
plt.imshow(bright_improve, cmap="gray")
plt.title("Brightness Improved")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
C = 50
bright_reduce = cv2.subtract(image, C)
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 3)
plt.imshow(bright_reduce, cmap="gray")
plt.title("Brightness Reduced")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
_, thresh = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 4)
plt.imshow(thresh, cmap="gray")
plt.title("Thresholding")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
negative = 255 - image
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 5)
plt.imshow(negative, cmap="gray")
plt.title("Negative")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
image_float = image.astype(float)
c = 255 / np.log(1 + np.max(image_float))
log_transform = c * (np.log(1 + image_float))
log_transform = np.array(log_transform, dtype=np.uint8)
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 6)
plt.imshow(log_transform, cmap="gray")
plt.title("Log Transformation")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
gamma = 0.5   # try <1 (bright) or >1 (dark)
gamma_corrected = np.array(255 * ((image / 255) ** gamma), dtype=np.uint8)
plt.figure(figsize=(20, 15))
plt.subplot(2, 4, 7)
plt.imshow(gamma_corrected, cmap="gray")
plt.title(f"Gamma (={gamma})")
plt.axis("off")

plt.tight_layout()
plt.show()
```
