```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sunset.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
if img is None:
    raise ValueError("Image not loaded. Check the file path!")
```

```python
# Contrast Stretching
min_val = np.min(img)
max_val = np.max(img)
contrast_stretched = ((img - min_val) / (max_val - min_val)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)
```

```python
# Histogram Equalization
hist_eq = cv2.equalizeHist(img)
```

```python
# Plot Images and Histograms
plt.figure(figsize=(15,10))
```

```python
# Original
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Low-Contrast Image")
plt.axis("off")
```

```python
plt.subplot(2,2,2)
plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram (Original)")
```

```python
# Contrast Stretched
plt.subplot(2,2,1)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis("off")
```

```python
plt.subplot(2,2,1)
plt.hist(contrast_stretched.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram (Contrast Stretched)")
```

```python
# Histogram Equalized
plt.subplot(2,2,1)
plt.imshow(hist_eq, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")
```

```python
plt.subplot(2,2,1)
plt.hist(hist_eq.ravel(), bins=256, range=[0,256], color='black')
plt.title("Histogram (Equalized)")
```

```python
plt.tight_layout()
plt.show()
```
